```python
# Credit to kenjiuno for the PCode.xml file
import xmltodict, json
xml = open("PCode.xml").read()
obj = xmltodict.parse(xml)
```

```python
import collections
instructions = {}
for instr in obj["PCode"]["Instr"]:
    new = {}
    for k in instr:
        name = k[1:] if k.startswith("@") else k
        value = instr[k]
        if type(value) == collections.OrderedDict:
            vnew = {}
            for vk in value:
                vname = vk[1:] if vk.startswith("@") else vk
                vvalue = value[vk]
                vnew[vname] = vvalue
            value = vnew
        new[name] = value
    instructions[new["name"]] = new
```

```python
traps = {}
for trap in obj["PCode"]["Syscall"]:
    new = {}
    for k in trap:
        name = k[1:] if k.startswith("@") else k
        value = trap[k]
        new[name] = value
    traps[new["name"]] = new
print(traps)
```

```python
json.dump(instructions, open("instructions.json", "w"), indent=4)
json.dump(traps, open("traps.json", "w"), indent=4)
```
